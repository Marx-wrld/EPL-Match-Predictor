### Creating predictors for our ML Model

```python
matches["venue_code"] =  matches["venue"].astype("category").cat.codes
## converting the home and away column into a numeric column 
## It will also convert it into a categorical data typ in pandas
## converting it from strings to categories and then the categories to numbers
```

```python
matches
```

```python
## we'll create a unique code for each opponents code
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches
```

```python
## we'll also check at the time, maybe certain teams play better t certain times of the game
## we're going to write an expression and this expression will replace the colon and minutes with nothing
## we'll again the hour to an integer because as you remember we need to be able to use a number as an input to an ML algorithm
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
## converts W to 1 meaning win and L to 0 meaning lose
```

```python
matches
```

### Creating Our Initial ML Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
## if we set a random state it means that if we run the random forest multiple times we get the same values
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
## fit is going to train a random forest model
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
## comparing our actual values and predicted values
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
```

```python
## create a crosstab using pandas
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
## we need to go back and revise our accuracy because what we care about more is predicting winning
from sklearn.metrics import precision_score
```

### Improving Precision with Rolling Averages

```python
grouped_matches = matches.groupby("team")
## This will create one dataframe for every squad in our data
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
## we took our original matches dataframe and grouped it by team which always creates one dataframe for each team 
## and then we apply the function to each of those dataframes to compute the rolling averages
### As you see below now we have the rolling_average for every match
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
## This will assign values from 0 t0 1316 to be our new indices
```

```python
matches_rolling
```

### Retraining our ML model

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] < '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
